```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("Maruti auto.csv")
months=df["Date "]
months=list(months)
for i in range(0,len(months)):
  if len(months[i]) == 8:
    months[i]=months[i][2:8]
  else:
    months[i]=months[i][3:9]
# months=list(set(months))
print(months)
df['OPEN '] = df['OPEN '].astype(float)
df['HIGH '] = df['HIGH '].astype(float)
df['LOW '] = df['LOW '].astype(float)
df['close '] = df['close '].astype(float)
df.head()
```

```python
df_stocks=df[["OPEN ","HIGH ","LOW ","close "]]
df_stocks.head()
df_stocks["months"]=months
```

```python
fig, ax = plt.subplots(1,1,figsize = (12,30))
# ages = df_ages[df_ages["date"] == year]
  # ages = ages.drop(['date','total'], 1)
stocks = df_stocks.set_index('months')
plt.title('stocks in Maruthi')
sns.heatmap(stocks, ax=ax)
```
